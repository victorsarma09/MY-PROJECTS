**5. MACHINE LEARNING MODEL SELECTION & MODEL TRAINING**

```python
#1. Splitting the Dataset into X and Y
X = df.drop(["winner"], axis=1)
y = df["winner"]
```

```python
X.head(4)
```

```python
#2. Converting categorical values into numerical values
X = pd.get_dummies(X, ["team1","team2", "toss_winner", "toss_decision", "result"], drop_first = True)
```

```python
#3. Converting the output data (categorical value) into numerical value
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
#4. Converting our data into a training set in order to create the model and testing set for evaluating the created model / Applying the Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
#5. Applying ML MODELS
#A. RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=200,min_samples_split=3,
                               max_features = "auto")

#Fitting the model
model.fit(x_train, y_train)

#Predicting the model
y_pred = model.predict(x_test)

#Evaluating the model
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
#B. SUPPORT VECTOR CLASSIFIER
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()

#Fitting the model
model.fit(x_train, y_train)

#Predicting the model
y_pred = model.predict(x_test)

#Evaluating the model
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
#C. DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

#Fitting the model
model.fit(x_train, y_train)

#Predicting the model
y_pred = model.predict(x_test)

#Evaluating the model
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
#D. GRADIENT BOOSTING CLASSIFIER
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()

#Fitting the model
model.fit(x_train, y_train)

#Predicting the model
y_pred = model.predict(x_test)

#Evaluating the model
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
#E. XGB CLASSIFIER
from xgboost import XGBClassifier
model = XGBClassifier()

#Fitting the model
model.fit(x_train, y_train)

#Predicting the model
y_pred = model.predict(x_test)

#Evaluating the model
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```
